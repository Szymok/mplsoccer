```python
# Step 1: Import necessary libraries
import matplotlib.pyplot as plt
from mplsoccer.pitch import Pitch

# Step 2: Create a Pitch object
pitch = Pitch(pitch_type='statsbomb',
              pitch_color='#22312b', line_color='#c7d5cc')

# Step 3: Create a figure and axes (we get a fig and ax by calling the draw method)
fig, ax = pitch.draw()

# Step 4: Show the plot
plt.show()
```

```python
# Step 1: Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
from mplsoccer.pitch import Pitch

# Step 2: Load some sample data (replace with your own data)
data = pd.DataFrame({'x': [60, 40, 30, 70], 'y': [40, 20, 60, 50]})

# Step 3: Create a Pitch object
pitch = Pitch(pitch_type='statsbomb',
              pitch_color='#22312b', line_color='#c7d5cc')

# Step 4: Create a figure and axes (we get a fig and ax by calling the draw method)
fig, ax = pitch.draw()

# Step 5: Plot the events on the pitch
sc = pitch.scatter(data['x'], data['y'], s=200, ax=ax)

# Step 6: Show the plot
plt.show()
```
