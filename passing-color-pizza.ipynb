```python
from urllib.request import urlopen

import matplotlib.pyplot as plt
from PIL import Image

from mplsoccer import PyPizza, add_image, FontManager
import soccerdata as sd

import pandas as pd

# Ustawienie opcji wyświetlania na 'None' spowoduje wyświetlenie wszystkich kolumn
pd.set_option('display.max_columns', None)
```

```python
font_normal = FontManager('https://github.com/google/fonts/raw/main/apache/robotomono/RobotoMono-Italic%5Bwght%5D.ttf')
font_italic = FontManager('https://github.com/google/fonts/raw/main/apache/robotoslab/RobotoSlab%5Bwght%5D.ttf')
font_bold = FontManager('https://github.com/google/fonts/raw/main/apache/maidenorange/MaidenOrange-Regular.ttf')
```

```python
URL = "https://raw.githubusercontent.com/andrewRowlinson/mplsoccer-assets/main/fdj_cropped.png"
fdj_cropped = Image.open(urlopen(URL))
```

```python
fbref = sd.FBref(leagues="ESP-La Liga", seasons=2023)
print(fbref.__doc__)
```

```python
player_match_stats = fbref.read_player_match_stats(stat_type="passing", match_id='872da7fb')
player_match_stats.head()
```

```python
specific_player_data = player_match_stats.xs('Jude Bellingham', level='player')
specific_player_data.head()
```

```python
# parameter list
params = [
        'PC', 'PA', 'PC%', 'PCS', 'PAS', 'PCS%', 'PCM', 'PAM', 'PCM%', 'PCL', 'PAL', 'PCL%'
]

# value list
# The values are taken from the excellent fbref website (supplied by StatsBomb)

values = specific_player_data[[('Total', 'Cmp'), ('Total', 'Att'), ('Total', 'Cmp%'), ('Short', 'Cmp'), ('Short', 'Att'), ('Short', 'Cmp%'), ('Medium', 'Cmp'), ('Medium', 'Att'), ('Medium', 'Cmp%'), ('Long', 'Cmp'), ('Long', 'Att'), ('Long', 'Cmp%')]].values.flatten().tolist()
# color for the slices and text
slice_colors = ["#1A78CF"] * 3 + ["#FF9300"] * 3 + ["#D70232"] * 3 + ["#d5d57e"] * 3
text_colors = ["#000000"] * 12

# instantiate PyPizza class
baker = PyPizza(
    params=params,                  # list of parameters
    background_color="#EBEBE9",     # background color
    straight_line_color="#EBEBE9",  # color for straight lines
    straight_line_lw=1,             # linewidth for straight lines
    last_circle_lw=0,               # linewidth of last circle
    other_circle_lw=0,              # linewidth for other circles
    inner_circle_size=20            # size of inner circle
)

# plot pizza
fig, ax = baker.make_pizza(
    values,                          # list of values
    figsize=(8, 8.5),                # adjust figsize according to your need
    color_blank_space="same",        # use same color to fill blank space
    slice_colors=slice_colors,       # color for individual slices
    value_colors=text_colors,        # color for the value-text
    value_bck_colors=slice_colors,   # color for the blank spaces
    blank_alpha=0.4,                 # alpha for blank-space colors
    kwargs_slices=dict(
        edgecolor="#F2F2F2", zorder=2, linewidth=1
    ),                               # values to be used when plotting slices
    kwargs_params=dict(
        color="#000000", fontsize=11,
        fontproperties=font_normal.prop, va="center"
    ),                               # values to be used when adding parameter
    kwargs_values=dict(
        color="#000000", fontsize=11,
        fontproperties=font_normal.prop, zorder=3,
        bbox=dict(
            edgecolor="#000000", facecolor="cornflowerblue",
            boxstyle="round,pad=0.2", lw=1
        )
    )                                # values to be used when adding parameter-values
)

# add title
fig.text(
    0.515, 0.975, "Jude Bellingham - Real Madrid", size=16,
    ha="center", fontproperties=font_bold.prop, color="#000000"
)

# add subtitle
fig.text(
    0.515, 0.953,
    "Passes vs. Real Betis | Season 2023-24",
    size=13,
    ha="center", fontproperties=font_bold.prop, color="#000000"
)

# add credits
CREDIT_1 = "data: viz fbref"
CREDIT_2 = "inspired by: @skszymon"

fig.text(
    0.99, 0.02, f"{CREDIT_1}\n{CREDIT_2}", size=9,
    fontproperties=font_italic.prop, color="#000000",
    ha="right"
)

# add text
fig.text(
    0.34, 0.925, "Total                           Short                             Medium                        Long", size=14,
    fontproperties=font_bold.prop, color="#000000"
)

# add rectangles
fig.patches.extend([
    plt.Rectangle(
        (0.31, 0.9225), 0.025, 0.021, fill=True, color="#1a78cf",
        transform=fig.transFigure, figure=fig
    ),
    plt.Rectangle(
        (0.462, 0.9225), 0.025, 0.021, fill=True, color="#ff9300",
        transform=fig.transFigure, figure=fig
    ),
    plt.Rectangle(
        (0.632, 0.9225), 0.025, 0.021, fill=True, color="#d70232",
        transform=fig.transFigure, figure=fig
    ),
    plt.Rectangle(
        (0.8, 0.9225), 0.025, 0.021, fill=True, color="#d5d57e",
        transform=fig.transFigure, figure=fig
    )
])

# add image
# ax_image = add_image(
#     fdj_cropped, fig, left=0.4478, bottom=0.4315, width=0.13, height=0.127
# )   # these values might differ when you are plotting

plt.show()
```
